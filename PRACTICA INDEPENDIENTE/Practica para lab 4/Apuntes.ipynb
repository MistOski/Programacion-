Creando un grafico de barras

```python
conteo = datos['Survived'].value_counts() #Cuenta cuantas veces aparece cada valor en la columna 'survived'

etiquetas = ['Fallecidos', 'Sobrevivientes'] #Se crean las etiquetas y se guardan en una varable

#Se crea el grafico
plt.bar(etiquetas, conteo.sort_index(), color=['red', 'green']) #conteo.sort_index() permite ordenar los fallecidos de 0 a 1, para que ellos vayan primero
plt.title('Supervivientes vs Fallecidos') #Se establece el titulo
plt.xlabel('Estado') #Se agrega el sub titulo para asegurarse de que se estan indicando los valores correctos en el lugar adecuado
plt.ylabel('Cantidad de personas')
plt.show()
```

Creacion de grafico circular

```python

# Contamos la cantidad de sobrevivientes y fallecidos
conteo = datos['Survived'].value_counts().sort_index()
# Etiquetas y colores
etiquetas = ['Fallecidos', 'Sobrevivientes']
colores = ['red', 'green']

# Graficamos
plt.pie(
    conteo,
    labels=etiquetas,
    colors=colores,
    autopct='%1.1f%%',    # Muestra el porcentaje con un decimal
    startangle=90,         # Comienza el grfico desde arriba
    explode=(0.05, 0)      # Separa un poco el primer segmento (fallecidos)
)

plt.title('Proporcin de Supervivientes y Fallecidos')
plt.axis('equal')  # Asegura que el crculo se vea bien redondo
plt.show()
```

Grafico de lineas

```python
# Contamos la cantidad de fallecidos (0) y sobrevivientes (1), ordenados por ndice
conteo = datos['Survived'].value_counts().sort_index()

# Etiquetas para el eje X
etiquetas = ['Fallecidos', 'Sobrevivientes']

# Graficamos una lnea
plt.plot(etiquetas, conteo, marker='o', linestyle='-', color='blue')

# Detalles del grfico
plt.title('Cantidad de Fallecidos y Sobrevivientes')
plt.xlabel('Estado')
plt.ylabel('Cantidad de personas')
plt.grid(True)
plt.show()
```
